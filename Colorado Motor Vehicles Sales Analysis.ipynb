```python
import pandas as pd
data=pd.read_csv('colorado.csv')
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
columns=data.columns
print(columns)
```

```python
#Importin necessary modules
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default='plotly_white'
import matplotlib.pyplot as plt
```

```python
fig=px.bar(data,x='county',y='sales',color='sales',title='Sales level over County')
fig.update_layout(xaxis_title='County',
                 yaxis_title='Sales Level')
fig.show()
```

So The highest sales level was in Araphoe County. Now let us take a look at the level of quarterly sales

```python
sales_by_quarter=data.groupby('quarter')['sales'].sum().reset_index()
print(sales_by_quarter)
labels=sales_by_quarter['quarter']
values=sales_by_quarter['sales']
colors=['gold', 'lightgreen', 'skyblue', 'lightcoral']
fig=go.Figure(data=[go.Pie(labels=labels,values=values)])
fig.update_layout(title_text='Sum of Sales by quarters',title_x=0.5)
fig.update_traces(hoverinfo='label+percent',textinfo='percent',textfont_size=20,marker=dict(colors=colors,line=dict(color='black',width=2)))
fig.show()
```

The highest sales occured in third quarter.
Let us take a look at the distribution of sales through the years.

```python
yearly_sales=data.groupby('year')['sales'].sum().reset_index().rename(columns={'sales':'Yearly_sale'})
data=data.merge(yearly_sales,on='year')
fig=px.scatter(data,x='Yearly_sale',y='sales',size='sales',color='sales')
fig.update_layout(title_text='Distribution of sales by year',xaxis_title='Years',yaxis_title='The size of sales')
fig.show()
```
